```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Independent Variable (Feature)

num_hours_studied = np.array([1,3,3,4,5,6,7,7,8,8,10])

# Dependent variable (Target)

exam_score = np.array([18,26,31,40,55,62,71,70,75,85,97])

# Plotting the data
plt.scatter(num_hours_studied, exam_score)
plt.xlabel("Num of hours studied")
plt.ylabel("Exam Score")
plt.title("Exam Score Data Set")
plt.show()
```

### Cost Function

```python
## function to calculate the cost

def compute_cost(x,y,w,b):
    m = x.shape[0]
    cost = 0
    
    for i in range(m):
        f_wb = w*x[i] + b
        cost += (f_wb - y[i])**2
    cost = cost/(2*m)
        
    return cost
```

### compute gradient
`compute_gradient` returns `dj_dw`,`dj_db`

```python
def compute_gradient(x,y,w,b):
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0
    
    for i in range(m):
        f_wb = w*x[i] + b
        dj_dw += (f_wb - y[i])*x[i]
        dj_db += (f_wb - y[i])
    
    return dj_dw/m, dj_db/m
```

### Gradient Descent
Now, since, we can compute cost, gradient, we'll implement `gradient descent` to find the optimal values of <strong>w and b</strong>

```python
def gradient_descent(x , y, w_in, b_in, alpha, num_iter, compute_cost, compute_gradient):
    
    w = w_in
    b = b_in
    
    for i in range(num_iter):
        dj_dw, dj_db = compute_gradient(x,y,w,b)
        
        #updating parameters w and b
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
        
    return w,b
```

```python
# initializing parameters
w_in = 0
b_in = 0
num_iter = 10000
alpha = 1.0e-2

w,b = gradient_descent(num_hours_studied, exam_score, w_in, b_in, alpha, num_iter, compute_cost, compute_gradient)

print(f" w and b are {w} , {b} with cost function J_wb = {compute_cost(num_hours_studied,exam_score,w,b)}")
```

 ### Comparing data points with our prediction model

```python
#our predictions
m = num_hours_studied.shape[0]
f_wb = np.zeros(m)
for i in range(m):
    f_wb[i] = w*num_hours_studied[i] + b
```

```python
# plot our model prediction
plt.plot(num_hours_studied, f_wb, c='b', label='Our Prediciton')

#plot the data points
plt.scatter(num_hours_studied, exam_score, marker = 'x', c='r', label='Actual Values')

plt.title("Exam Score Data Set")
plt.ylabel("Exam Score")
plt.xlabel("No of hours studied")
plt.legend()
plt.show()
```
